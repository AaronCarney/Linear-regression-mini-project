```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
from scipy import stats
```

```python
# Load Data

hd = pd.read_csv ('HD.csv')
amd = pd.read_csv('AMD.csv')
wmt = pd.read_csv('WMT.csv')

# The average price is considered the average of the High, Low and Close
# Nothing fancy required

hd['Average'] = (hd['High'] + hd['Low'] + hd['Close']) / 3
amd['Average'] = (amd['High'] + amd['Low'] + amd['Close']) / 3
wmt['Average'] = (wmt['High'] + wmt['Low'] + wmt['Close']) / 3

# Data is already clean, dropp unnecessary columns
hd = hd.drop(columns=['High','Low','Close'])
amd = amd.drop(columns=['High','Low','Close'])
wmt = wmt.drop(columns=['High','Low','Close'])

# Date isn't in a workable datatype so
hd['Date'] = pd.to_datetime(hd.Date)
amd['Date'] = pd.to_datetime(amd.Date)
wmt['Date'] = pd.to_datetime(wmt.Date)
```

```python
# LinReg and plotting for HD

# remove the headers
data = hd[:].values

# Setup subplots
fig, ax = plt.subplots()

# Separating arrays
dates = [x[0] for x in data]
price = [x[1] for x in data]

# linregress and bestfit line
bestfit = stats.linregress(range(len(price)),price)
firstorder = str(round(bestfit[0],2)) + "x + " + str(round(bestfit[1],2)) 

ax.plot(range(len(price)), price)
ax.plot(range(len(price)), np.poly1d(np.polyfit(range(len(price)), price, 1))(range(len(price))), '--',label=firstorder)
ax.plot(range(len(price)), np.poly1d(np.polyfit(range(len(price)), price, 2))(range(len(price))), '--',label="Second Order")

# Set date tick labels and legend
plt.xticks([0,60,120,180,240], ['2003', '2008', '2013','2018','2023']) 
plt.xlabel('Year (tick on April 1st of the year)')
plt.ylabel('Price ($)')
plt.legend()

# Display plot
plt.show()
```
